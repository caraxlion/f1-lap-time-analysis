# Formula One Lap Analysis Visualization

```python
import fastf1
from fastf1 import plotting
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import os
```

```python
#define the cache directory
cache_dir = 'cache'

#create the cache directory if it does not exist
if not os.path.exists(cache_dir):
    os.makedirs(cache_dir)

# Enable FastF1 cache
fastf1.Cache.enable_cache(cache_dir)
```

```python
# Load a race session (e.g. 2023 Monaco GP - race)
session = fastf1.get_session(2023, 'Monaco', 'R')
session.load()
```

```python
#Get the lap data for a specific driver 
# example: Max Verstappen
laps = session.laps.pick_drivers('VER')
fastest_lap = laps.pick_fastest()

telemetry = fastest_lap.telemetry.add_distance()
```

```python
# Plot the speed vs. distance
plt.figure(figsize=(10, 5))
plt.plot(telemetry['Distance'], telemetry['Speed'])
plt.title("Verstappen - Fastest Lap Speed (Monaco 2023)")
plt.xlabel("Distance (m)")
plt.ylabel("Speed (km/h)")
plt.grid()
plt.savefig("../visuals/ver_fastest_lap_monaco.png")
plt.show()
```
